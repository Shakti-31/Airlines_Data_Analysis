```python
ax = sns.countplot(x='airline' , data =data)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x='departure_time' , data =data, palette='rocket')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x='arrival_time' , data =data, palette='rocket')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
stops = data['stops'].value_counts(normalize=True)
print(stops)

plt.figure(figsize=(5,4))
plt.pie(stops, labels=stops.index, autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```

```python
class_counts = data['class'].value_counts(normalize=True)
print(class_counts)

plt.figure(figsize=(5,4))
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x='source_city' , data =data, palette='rocket')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x='destination_city' , data =data, palette='viridis')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sns.scatterplot(x='days_left', y='price' , data=data)
plt.title('Days Left vs Price')
plt.show()
```

```python
sns.scatterplot(x='days_left', y='duration' , data=data)
plt.title('Days Left vs duration')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='airline', data=data, palette='viridis')
plt.title('airline and price')
plt.xlabel('airline')
plt.ylabel('price')
```

```python
data.groupby('days_left')['duration'].mean().sort_values(ascending=False)
```
